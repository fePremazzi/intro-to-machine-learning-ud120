- Of these three individuals (Lay, Skilling and Fastow), who took home the most money (largest value of “total_payments” feature)?

```python
andrew = enron_data["FASTOW ANDREW S"]["total_payments"]
jeff = enron_data["SKILLING JEFFREY K"]["total_payments"]
kenn = enron_data["LAY KENNETH L"]["total_payments"]
maior = "FASTOW ANDREW S"
maiormoney = andrew
if jeff > maiormoney :
    maior = "SKILLING JEFFREY K"
    maiormoney = jeff
if kenn > maiormoney :
    maior = "LAY KENNETH L"
    maiormoney = kenn

print maior, " took the most money with the amount: $", maiormoney
```
